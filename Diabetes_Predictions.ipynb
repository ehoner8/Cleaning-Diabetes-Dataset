```python
import re
```

```python
def convert_age_to_int(age_range):
    expression = "^\[([0-9]*)-([0-9]*)\)$"
    matches = re.search(expression, age_range)
    age_number = (int(matches.group(1)) + int(matches.group(2))) / 2
    return age_number
```

```python
print(convert_age_to_int("[50-60)"))
```

```python
df["ages"] = df["age"].apply(convert_age_to_int)
```
